```python
import apache_beam as beam

def print_first_5_records(record):

    if print_first_5_records.count < 5:
        print(record)
        print_first_5_records.count += 1

print_first_5_records.count = 0

with beam.Pipeline() as p:

    parquet_lines = (
        p
        | 'ReadParquet' >> beam.io.ReadFromParquet('/content/yellow_tripdata_2020-03.parquet')
    )


    parquet_lines | beam.Map(print_first_5_records)

```

```python
def collect_records(record):
    if collect_records.count < 5:
        collect_records.records.append(record)
        collect_records.count += 1

collect_records.count = 0
collect_records.records = []

with beam.Pipeline() as p:
    parquet_lines = (
        p
        | 'ReadParquet' >> beam.io.ReadFromParquet('/content/yellow_tripdata_2020-03.parquet')
    )
    parquet_lines | beam.Map(collect_records)
labels = [f'Record {i+1}' for i in range(len(collect_records.records))]
values = [str(record) for record in collect_records.records]

plt.bar(labels, values)
plt.xlabel('Record Number')
plt.ylabel('Record Value')
plt.title('First 5 Records from Parquet File')
plt.show()
```
